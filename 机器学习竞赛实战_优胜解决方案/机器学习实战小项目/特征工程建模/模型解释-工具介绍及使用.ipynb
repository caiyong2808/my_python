### Permutation Importance

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
data = pd.read_csv('data/FIFA 2018 Statistics.csv')  # 足球赛事数据集
data.head()
```

怎么知道哪些特征影响多大

破坏特征：比如把Goal Scored上下的顺序打乱，其它不变，然后比较破坏前和后的区别

如果前 ≈ 后，说明这个特征对结果没影响
如果后 ＜ 前，说明这个特征对结果影响起正作用
如果后 ＞ 前，说明这个特征对结果影响起负作用（一般不会出现）

```python
y = (data['Man of the Match'] == 'Yes')  # 转换标签
y[:5]
```

```python
feature_names = [i for i in data.columns if data[i].dtype in [np.int64]]
X = data[feature_names]
X.head()
```

```python
train_X, val_X,train_y,val_y = train_test_split(X, y,random_state=1)
my_model = RandomForestClassifier(random_state=0).fit(train_X,train_y)
```

```python
import eli5 # pip install eli5
from eli5.sklearn import PermutationImportance

perm = PermutationImportance(my_model,random_state=1).fit(val_X, val_y)

eli5.show_weights(perm, feature_names=val_X.columns.tolist())
```

### Partial Dependence Plots

特征重要性展示了每个特征发挥的作用情况，partial dependence plots 可以展示一个特征怎么影响预测结果

模型建立完成后进行使用，概述如下：
<ul>
    <li>首先选中一个样本数据，此时想观察Ball Possession列对结果的影响。
    <li>保证其它特征列不变，改变当前观察列的值，例如悬着40%，50%，60%分别进行预测，得到各自的结果。
    <li>对比结果就能知道当前列（Ball Possession）对结果的影响情况。

```python
import matplotlib.pyplot as plt
from pdpbox import pdp, get_dataset, info_plots  # pip install pdpbox
pdp_goals = pdp.pdp_isolate(model=my_model,
                            dataset=val_X, 
                            model_features=feature_names,
                            feature='Goal Scored')  # 观察的特征

pdp.pdp_plot(pdp_goals, 'Goal Scored')
plt.show()
```

y轴表示预测结果的变化对比于基本模型，由于在观察时不可能只看一个样本数据，肯定要选择多个样本数据，蓝色区域表示的是置信度

```python
feature_to_plot = 'Distance Covered (Kms)'

rf_model = RandomForestClassifier(random_state=0).fit(train_X, train_y)

pdp_dist = pdp.pdp_isolate(model=rf_model, dataset=val_X,
                          model_features=feature_names,
                          feature=feature_to_plot)

pdp.pdp_plot(pdp_dist,feature_to_plot)
plt.show()
```

### 双特征观察

### SHAP Values
可以直观的展示每一个特征对结果走势的影响

```python
row_to_show = 5
data_for_prediction = val_X.iloc[row_to_show]
data_for_prediction_array = data_for_prediction.values.reshape(1, -1)

my_model.predict_proba(data_for_prediction_array)  # 能赢的概率为70%
```

导入shap工具包可能出现问题，在Anaconda3\Lib\site-packages\numpy\lib\arraypad.py中添加下面两个函数到_as_pairs和_as_pairs之间（956行）保存，重新加载即可，utf-8格式

```python
def _normalize_shape(ndarray, shape, cast_to_int=True):
    ndims = ndarray.ndim
    
    if shape is None:
        return ((None, None),) * ndims
    
    shape_arr = np.asarray(shape)
    
    try:
            shape_arr = np.broadcast_to(shape_arr, (ndims, 2))
    except ValueError:
        fmt = "Unable to create correctly shaped tuple from %s"
        raise ValueError(fmt % (shape,))
    
    if cast_to_int is True:
        shape_arr = np.round(shape_arr).astype(int)
        
    return tuple(tuple(axis) for axis in shape_arr.tolist())


def _validate_lengths(narray, number_elements):
    normshp = _normalize_shape(narray, number_elements)
    for i in normshp:
        chk = [1 if x is None else x for x in i]
        chk = [1 if x >= 0 else -1 for x in chk]
        if (chk[0] > 0) or (chk[1] < 0):
            fmt = "%s cannot contain negative values."
            raise ValueError(fmt % (number_elements,))
    return normshp
```

```python
import shap  # pip install shap

#实例化
explainer = shap.TreeExplainer(my_model)

#计算
shap_values = explainer.shap_values(data_for_prediction)
```

返回的SHAP values中包括了negative和positive两种情况，通常选择一种（positive）即可

```python
shap.initjs()
# shap_values[1]指定为positive，0为negative
shap.force_plot(explainer.expected_value[1], shap_values[1], data_for_prediction)
```

0.70表示赢得概率，红色表示能够朝1得方向发展得，蓝色表示能够朝0得方向发展的；

即红色的值越大越容易赢，蓝色值越大越容易输；

多个深红色小块之间，区域越大，表示影响结果的能力越大。蓝色同理。如图中绿框区域

### Summary Plots

```python
explainer = shap.TreeExplainer(my_model)

shap_values = explainer.shap_values(val_X)  # 选择全部样本

shap.summary_plot(shap_values[1], val_X)
```

Y轴是特征，X轴是正负，正的表示特征对结果起促进作用，负的表示特征对结果起抑制作用。右边Y轴表示越高，对结果影响越大。

如上图中：
    Goal Scored 左边区域的蓝色：进球越少就越抑制赢球这一结果。
    Goal Scored 右边区域的红色：进球越多就越促进赢球这一结果。
