```python
import skimage
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from skimage import io

%matplotlib inline
```

## 图像shape

```python
cat = io.imread('data/cat.jpg')
dog = io.imread('data/dog.jpg')
df = pd.DataFrame(['Cat', 'Dog'], columns=['Image'])
```

```python
cat  # 0-255，越小的值表示越暗，越大则越凉
```

```python
print(cat.shape, dog.shape)
fig = plt.figure(figsize=(8,4))
ax1 = fig.add_subplot(1,2,1)
ax1.imshow(cat)
ax2 = fig.add_subplot(1,2,2)
ax2.imshow(dog)
```

三维的数据，250是高，300的宽，3是颜色通道红绿蓝。

```python
# 分别使用3个颜色通道
dog_r = dog.copy()
dog_r[:,:,1] = dog_r[:,:,2] = 0
dog_g[:,:,0] = dog_r[:,:,2] = 0
dog_b = dog.copy()
dog_b[:,:,0] = dog_g[:,:,1] = 0

plot_image = np.concatenate((dog_r,dog_g,dog_b),axis=1)
plt.figure(figsize=(10,4))
plt.imshow(plot_image)
```

## 灰度图

```python
from scipy import misc
cat_gray = misc.imread('data/cat.jpg', mode="L") 
dog_gray = misc.imread('data/dog.jpg', mode="L") 

fig = plt.figure(figsize=(8,4))
ax1 = fig.add_subplot(2,2,1)
ax1.imshow(cat_gray,cmap="gray")
ax2 = fig.add_subplot(2,2,2)
ax2.imshow(dog_gray,cmap="gray")
```

io.imread的效果不一定能变成灰度图，可以暂时用这个方法，操作一样，只是读取图片的代码不一样
